```python
import pandas as pd
import numpy as np
```

```python
t1 = pd.read_excel('/content/Подразделение1.xlsx', sheet_name="Часы 01-23")
t2 = pd.read_excel('/content/Подразделение1.xlsx', sheet_name="Часы 02-23")
t3 = pd.read_excel('/content/Подразделение2.xlsx', sheet_name="Часы 01-23")
t4 = pd.read_excel('/content/Подразделение2.xlsx', sheet_name="Часы 02-23")
```

1. Общее кол-во сотрудников из выбранного подразделения, за выбранный период по всем подрядчикам. Динамика по дням.  
2. Соотношение долей работающих сотрудников (в процентах) разных подразделений за выбранный период.
3. Общее количество сотрудников, которые работали в выбранный день.
P.S. для решения задачи нужно использовать одно подключение.

```python
t1.head()
```

```python
df1 = t1.merge(t2, on = ['Компания', 'Фио', 'Подразделение'], how = 'outer')
df1.head()
```

```python
df1 = df1.melt(id_vars=['Компания', 'Фио', 'Подразделение'], var_name='Date', value_name='hours').fillna(0)
```

```python
df1.head()
```

```python
df2 = t3.merge(t4, on = ['Компания', 'Фио', 'Подразделение'], how = 'outer')
df2.head()
```

```python
df2 = df2.melt(id_vars=['Компания', 'Фио', 'Подразделение'], var_name='Date', value_name='hours').fillna(0)
```

```python
df = pd.concat([df1, df2])
```

```python
df.to_excel("excelsior.xlsx")
```

```python
df.to_csv("excelsior.csv")
```
